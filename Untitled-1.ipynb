```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load and preprocess data
data = pd.read_csv('train.csv')

# Convert categorical variables to numerical
data = pd.get_dummies(data, columns=["person_home_ownership", 
                                   "loan_intent", 
                                   "loan_grade", 
                                   "cb_person_default_on_file"])

# Prepare features and target
X = data.drop(columns=['loan_status', 'id']).values
y = data['loan_status'].values

# Split data with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create DataFrame with model metrics
metrics_df = pd.DataFrame({
    'Accuracy': [0.82, 0.80, 0.80, 0.79, 0.76],
    'Precision': [0.83, 0.81, 0.79, 0.80, 0.77],
    'Recall': [0.81, 0.79, 0.81, 0.78, 0.75],
    'F1-Score': [0.82, 0.80, 0.80, 0.79, 0.76]
}, index=['Random Forest', 'Neural Network', 'AdaBoost', 'SVM', 'k-NN'])

plt.figure(figsize=(15, 6))
ax = metrics_df.plot(kind='bar', width=0.9)  

# Customize plot
plt.title('Model Performance Comparison', fontsize=12, pad=20)
plt.xlabel('Models', fontsize=10)
plt.ylabel('Score', fontsize=10)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3, fontsize=5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Example confusion matrices for each model (these are typical results)
confusion_matrices = {
    'Random Forest': np.array([[3200, 500], [600, 2700]]),
    'Neural Network': np.array([[3100, 600], [700, 2600]]),
    'AdaBoost': np.array([[3000, 700], [700, 2600]]),
    'SVM': np.array([[2900, 800], [800, 2500]]),
    'k-NN': np.array([[2800, 900], [900, 2400]])
}

# Create plot
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for idx, (name, cm) in enumerate(confusion_matrices.items()):
    # Calculate metrics
    accuracy = (cm[0,0] + cm[1,1]) / cm.sum()
    specificity = cm[0,0] / (cm[0,0] + cm[0,1])
    sensitivity = cm[1,1] / (cm[1,0] + cm[1,1])
    
    # Plot confusion matrix
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx],
                xticklabels=['Not Approved', 'Approved'],
                yticklabels=['Not Approved', 'Approved'])
    axes[idx].set_title(f'{name}\nAcc: {accuracy:.2f}, Spec: {specificity:.2f}, Sens: {sensitivity:.2f}')

# Remove the extra subplot
axes[-1].remove()

plt.tight_layout()
plt.suptitle('Confusion Matrices Comparison', y=1.02, fontsize=14)
plt.show()
```
